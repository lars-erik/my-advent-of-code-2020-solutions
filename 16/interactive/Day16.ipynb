```
var inputLines = System.IO.File.ReadAllLines("./input.txt");

inputLines
```

```
var ranges = new List<(string name, (int from, int to) a, (int from, int to) b)>();
foreach(var line in inputLines)
{
    if (line == "") break;

    var part2 = line.Split(":")[1];
    var parts = part2.Trim().Split(" or ");
    var sets = parts.Select(x => {
        var values = x.Split("-").Select(y => Convert.ToInt32(y)).ToArray();
        return (from: values[0], to: values[1]);
    }).ToArray();
    ranges.Add((name: line.Split(":")[0], a: sets[0], b: sets[1]));
}

var others = inputLines.Where(x => x.Length > 0 && x.IndexOf(":") == -1)
     .Select(x => 
        x.Trim().Split(",")
            .Select(y => Convert.ToInt32(y))
            .Select((x, i) => (x:x, i:i, ranges:ranges.Where(r => (x >= r.a.from && x <= r.a.to) || (x >= r.b.from && x <= r.b.to))))
    );
```

```
// Answer

var invalidValues = others
    .SelectMany(x => x)
    .Where(x => x.ranges.Count() == 0);
    
invalidValues.Sum(x => x.x)
```

```
var validTickets = others
    .Where(x => x.All(y => y.ranges.Count() > 0));

validTickets
```
